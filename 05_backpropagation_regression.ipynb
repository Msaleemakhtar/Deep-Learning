```python
import numpy as np

def initialize_parameters(layer_dims):
    # Setting random seed for reproducibility
    np.random.seed(3)

    # Initializing an empty dictionary to store parameters
    parameters = {}

    # Getting the number of layers in the neural network
    L = len(layer_dims)

    # Looping through each layer (except the input layer)
    for l in range(1, L):


        # Initializing weights for layer l with dimensions (layer_dims[l-1], layer_dims[l])
        parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l])) * 0.1

        # Initializing biases for layer l with dimensions (layer_dims[l], 1)
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))

    return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
import numpy as np

def linear_forward(A_prev, W, b):
    # Compute the linear transformation Z = W^T * A_prev + b
    Z = np.dot(W.T, A_prev) + b



    # A_prev: Activations from the previous layer
    # W: Weight matrix for the current layer
    # b: Bias vector for the current layer

    return Z
```

```python
# Forward Prop
def L_layer_forward(X, parameters):
  # Initialize the input activations A to be X
  A = X

  # Calculate the number of layers in the neural network
  L = len(parameters) // 2


  # Loop through each layer of the neural network
  for l in range(1, L+1):
    # Set the activations of the previous layer as A_prev
    A_prev = A
    # Retrieve the weight matrix W and bias vector b for the current layer
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]

    # Print out the activations of the previous layer, weight matrix, and bias vector for the current layer
    print("A"+str(l-1)+": ", A_prev)
    print("W"+str(l)+": ", Wl)
    print("b"+str(l)+": ", bl)
    print("--"*20)

    # Perform linear transformation for the current layer
    A = linear_forward(A_prev, Wl, bl)

    # Print out the activations of the current layer
    print("A"+str(l)+": ", A)
    print("**"*20)

  # Return the activations of the output layer (A) and the activations of the previous layer (A_prev)
  return A, A_prev
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]

# Parameter initialization
parameters = initialize_parameters([2,2,1])


y_hat,A1 = L_layer_forward(X, parameters)


```

```python
X = df[['cgpa', 'profile_score']].values[1].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[1][0]

y_hat,A1 = L_layer_forward(X, parameters)
```

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
# epochs implementation

parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):

  Loss = []

  for j in range(df.shape[0]):

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = df[['lpa']].values[j][0]

    # Parameter initialization


    y_hat,A1 = L_layer_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

parameters
```
